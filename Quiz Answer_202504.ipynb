# a)Load CSV File

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
df=pd.read_csv(r"C:\Users\Office\Desktop\Student_Performance.csv")
print(df)
```

# b) Apply Linear Regression on the variables

```python
#X1 vs X6

x= df.X1
y=df.X6
x=sm.add_constant(x)
lr_model_X1 = sm.OLS(y,x).fit()

print(lr_model_X1.summary())
```

```python
#X2 vs X6

x= df.X2
y=df.X6
x=sm.add_constant(x)
lr_model_X2 = sm.OLS(y,x).fit()

print(lr_model_X2.summary())
```

```python
#X4 vs X6

x= df.X4
y= df.X6
x=sm.add_constant(x)
lr_model_X4 = sm.OLS(y,x).fit()

print(lr_model_X4.summary())
```

# c) Plot Scatter Plot with regression line

```python
#X1 vs X6

plt.figure()
plt.scatter(df.X1, df.X6)
plt.xlabel('Hours Studied')
plt.ylabel('Performance Index')
```

```python
# Plotting the data points and regression line
X_1=sm.add_constant(df.X1)
plt.scatter(df.X1, df.X6, color='blue', label='Observed Data')
plt.plot(df.X1, lr_model_X1.predict(X_1), color='red', label='OLS Regression Line')
plt.xlabel('Hours Studied')
plt.ylabel('Performance Index')
plt.title('Regression Line')
plt.legend()
plt.show()
```

# d) Apply multiple linear regression model

```python
Y= df['X6']
X_Total = df[['X1','X2','X3','X4','X5']]
X_Total =sm.add_constant(X_Total)
lr_model_MultipleR = sm.OLS(Y,X_Total).fit()

print(lr_model_MultipleR.summary())
```

```python
Y_Total=lr_model_MultipleR.predict(X_Total)
print("Predict of Performance Index")
print(Y_Total)
```

```python
X1 = 7
X2 = 90
X3= 1
X4 = 4
X5 = 1
Performance= lr_model_MultipleR.predict([1, X1, X2, X3, X4, X5])
print(Performance)
```
