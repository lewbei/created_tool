```python
import requests
import json

def get_doi_details(doi):
    url = f"https://api.crossref.org/works/{doi}"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json().get("message", {})
        # Remove the "reference" field if it exists
        data.pop("reference", None)
        return data
    else:
        print(f"Error: DOI {doi} returned status code {response.status_code}")
        return {}

doi = "10.1109/ICCV.2001.937709"
details = get_doi_details(doi)

if details:
    print(json.dumps(details, indent=2))  # Print details without "reference"
else:
    print("No details found for DOI:", doi)
```
